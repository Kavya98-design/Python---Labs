```python
#Maps with Marker
df_incidents = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_incidents.head()
```

```python
df_incidents.shape
```

```python
#consider first 100 incidents
df_incidents = df_incidents.iloc[0:100]
df_incidents
```

```python
df_incidents.shape
```

```python
#Latitude & Logitude of sanfarnsisco
latitude=37.77
longitude=-122.42
sanfran_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sanfran_map
```

```python
#create a feature group
incidents = folium.map.FeatureGroup()
incidents
```

```python
#loop through each lat & long and add child to each incidents in the form of circle marker
for lat,lng, in zip(df_incidents.Y,df_incidents.X):
    incidents.add_child(folium.features.CircleMarker([lat,lng],radius=5,color='yellow',fill=True,fill_color='blue',fill_opacity=0.6))
```

```python
#add all those markers for each incidents(child) to sanfran map
sanfran_map.add_child(incidents)
```

```python
#Add popup text
latitudes = list(df_incidents.Y)
longitudes=list(df_incidents.X)
label = list(df_incidents.Category)
for lat,lng, label, in zip(latitudes, longitudes, label):
    folium.Marker([lat,lng],popup=label).add_to(sanfran_map)
```

```python
sanfran_map
```

```python
sanfran_map.add_child(incidents)
```

```python
#To avoild marker congestion
sanfran_map = folium.Map(location = [latitude,longitude],zoom_start=12)
sanfran_map
```

```python
for lat,lng, label, in zip(df_incidents.Y,df_incidents.X,df_incidents.Category):
    folium.features.CircleMarker([lat,lng],radium=5,color='green',fill=True,fill_color='seagreen',popup=label,fill_opacity=0.6).add_to(sanfran_map)
    
```

```python
sanfran_map
```

```python
#to represent cluster by no of crimes
from folium import plugins
sanfran_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sanfran_map
```

```python
#create cluster group
incidents=plugins.MarkerCluster().add_to(sanfran_map)
incidents
```

```python
for lat,lng,label,in zip(df_incidents.Y,df_incidents.X,df_incidents.Category):
    folium.Marker(location=[lat,lng],popup=label,icon=None).add_to(incidents)
```
